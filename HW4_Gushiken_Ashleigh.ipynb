```python
import matplotlib.pyplot as plt

marks = {'Andy':88, 'Amy':66, 'James':90, 'Jules':55, 'Arthur':77}

plt.bar(marks.keys(), marks.values())
plt.title('Student Grades')
plt.xlabel('Students')
plt.ylabel('Grades')
plt.show()

plt.pie(marks.values(), labels=marks.keys(), autopct='%1.1f%%')
plt.title('Student Grades')
plt.legend(title='Students', loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.tan(x)
y2 = np.sqrt(x)
y3 = np.random.rand(5)
y4 = np.random.normal(size=100)
y5 = np.random.uniform(size=100)

fig, axs = plt.subplots(2, 3, figsize=(10, 6))

axs[0, 0].plot(x, y1)
axs[0, 0].plot(x, y2)
axs[0, 0].set_title('Multiple Lines')
axs[0, 0].legend()

axs[0, 1].bar(['A', 'B', 'C', 'D'], [10, 20, 30, 40])
axs[0, 1].set_title('Bar Chart')

axs[0, 2].pie(y3, labels=['A', 'B', 'C', 'D', 'E'])
axs[0, 2].set_title('Pie Chart')

axs[1, 0].plot(y4, linestyle='--', color='r')
axs[1, 0].grid(which='major', linestyle='-', linewidth='0.5', color='gray')
axs[1, 0].set_title('Grid')

axs[1, 1].scatter(x, y5, s=20, c='b')
axs[1, 1].set_title('Scatter Plot')

axs[1, 2].hist(y4, bins=10, alpha=0.5, edgecolor='black', color='green')
axs[1, 2].set_title('Histogram')

fig.suptitle('Subplots Example')

plt.show()
```
